# One hot encoding :

    One hot encoding is one method of converting categorical data to prepare it for an algorithm and get a better prediction. With one-hot, we convert each categorical value into a new categorical column and assign a binary value of 1 or 0 to those columns. Each integer value is represented as a binary vector. All the values are zero, and the index is marked with a 1.
    
    One hot encoding is useful for data that has no relationship to each other. Machine learning algorithms treat the order of numbers as an attribute of significance

```python
# Here i used get dummies function from pandas library and found that it is the better way to find the encoded values. it is converting categorical values to binary values. 

df_One_hot_encode = pd.get_dummies(df["neighbourhood_group"], prefix="Group")
```
